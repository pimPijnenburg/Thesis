```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, t
import os
```

```python
x = np.linspace(-4, 4, 100000)
gaussian = norm.pdf(x, 0, 1)
t = t.pdf(x, df = 1)
```

```python
sns.set_style('darkgrid')
plt.rcParams.update({'font.size': 20})

plt.figure(figsize = (18, 10))
sns.lineplot(x = x, y = gaussian, label = 'Gaussian distribution', linewidth = 2)
sns.lineplot(x = x, y = t, label = 't-Distribution with df = 1', linewidth = 2)

plt.title('Comparison of Gaussian and Student t-distribution Tails', fontsize = 20, pad=20)
plt.xlabel('x')
plt.ylabel('Probability Density')

plt.tight_layout()

output_path = '/Users/pimpijnenburg/Desktop/Thesis/USTC_SmokeRS_dataset/data/generated_images'
plot_output = os.path.join(output_path, 'gaussian_vs_t.png')
plt.savefig(plot_output)

plt.show()
```
